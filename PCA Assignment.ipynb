# Handling Outliers

```python
def Outliers(data,columns):
    for col in columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        IQR = q3 - q1
        Low_ext = q1 - (1.5*IQR)
        Upp_ext = q3 + (1.5*IQR)
        df[col] = np.where(df[col]>Upp_ext,Upp_ext,np.where(df[col]<Low_ext,Low_ext,df[col]))
```

```python
Outliers(df,df.iloc[:,2:6])
```

```python
Outliers(df,df.iloc[:,9:12])
```

# Dimensionality Reduction with PCA

```python
# Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(df.drop('Type',axis=1))
scaled_data
```

```python
# Implementaion of PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca_data = pca.fit_transform(scaled_data)
```

```python
# Scree Plot
plt.plot(range(1,len(pca.explained_variance_ratio_)+1),pca.explained_variance_ratio_,marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
# Cumulative Explained Variance
cumulative_variance = pca.explained_variance_ratio_.cumsum()
```

```python
plt.plot(range(1,len(cumulative_variance)+1),cumulative_variance,marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('No. of Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
# Transform original data into Principal components
pca1 = PCA(n_components=4)
optimal_pca_data = pca1.fit_transform(scaled_data)
pca_df = pd.DataFrame(optimal_pca_data, columns=[f'PC{i+1}' for i in range(4)])
```
