```python
def outlier(data,column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    IQR = q3-q1
    lower = q1-(1.5*IQR)
    upper = q3+(1.5*IQR)
    return(lower,upper)
```

```python
for i in df.select_dtypes(exclude='object').columns:
    lower,upper=outlier(df,i)
    print(i,'low_ext=',lower,' , upp_ext=',upper)
```
